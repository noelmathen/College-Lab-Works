```python
#merging two dictionaries using update
dic1={1:'a',2:'b'}
dic2={3:'c',4:'d'}
dic1.update(dic2)
print(dic1)
```

```python
#merging two dictionaries using ** operator
dic1={1:'a',2:'b'}
dic2={3:'c',4:'d'}
print({**dic1,**dic2})
```

```python
#merging two dictionaries using copy() and update()
dic1={1:'a',2:'b'}
dic2={3:'c',4:'d'}
dic3=dic2.copy()
print(dic3)
dic3.update(dic1)
print(dic3)
```

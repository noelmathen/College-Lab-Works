```python
#fibonacci series
number=int(input("Enter the limit: "))
num1, num2 = 0, 1
print(num1)
print(num2)
next=num1+num2
print(next)
for i in range(number):
  num1=num2
  num2=next
  next=num1+num2
  print(next)
```

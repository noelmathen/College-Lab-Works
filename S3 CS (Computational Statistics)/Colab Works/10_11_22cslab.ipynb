```python
#4. Python Program to Find the Square Root
import cmath as m
num=int(input("Enter the number: "))
print("The square root of ", num, "using sqrt function is ", m.sqrt(num))
print("The square root of ", num, "without function is ", int(num**0.5))
```

```python
#5. Print all prime numbers with in an interval
lb=int(input("Enter the lower limit: "))
ub=int(input("Enter the upper limit: "))
print("The prime numbers within ", lb, "and", ub, "is: ", end="")
for i in range(lb,ub+1):
    if i>1:
        for j in range(2,i):
            if(i%j==0):
                break
        else:
            print(i,end=" ")
```

```python
#6. Search an element in a list
lst=eval(input("Enter your list(any datatype): "))
search=eval(input("Enter the element to be searched: "))
if search in lst:
    print(search, "is present the list")
else:
    print(search, "is not present the list")
```

```python
#7. Input a list of n numbers. Calculate and display the average of numbers. Also display the square of each value in the list.
lst=eval(input("Enter your list of numbers: "))

#using for loop
sum=0
for i in lst:
    sum += i
foravg=sum/len(lst)
print("The average using for loop is: ",foravg)

#using other functions
funcavg=(sum(lst)/len(lst))
print("The average using functions is: ", funcavg)
```
